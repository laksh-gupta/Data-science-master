```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
## ploting line y = mx+c and f=x^2

x = np.arange(20)
y1 = x**2
y2 = 2*x+3
print(x)
print(y1)
print(y2)

plt.plot(x,y1,color="red")
plt.show()

plt.plot(x,y2,color="green")
plt.show()

plt.plot(x,y1,color="red",label="apple",marker=".")
plt.plot(x,y2,color="green",label="kiwi",linestyle="dashed")
plt.xlabel("time")
plt.ylabel("price")
plt.title("fruit price over time")
plt.legend()
plt.show()
```

```python
prices=np.array([1,2,3,4])**2
print(prices)
plt.plot(prices)
```

## Scatter plots

```python
## to change size
plt.figure(figsize=(10,10))

plt.scatter(x,y1,color="red",label="apple",marker=".")
plt.scatter(x,y2,color="green",label="kiwi",marker="^")
plt.xlabel("time")
plt.ylabel("price")
plt.title("Scatter plot-fruit prices over time")
plt.legend()
plt.show()
```

## Bar Graph

```python
x_codr = np.array([0,1,2])*2
plt.bar(x_codr,[10,20,15],width=0.5,label="Curr year",tick_label=["gold","platinum","silver"])
plt.bar(x_codr+0.5,[12,22,13],width=0.5,label="Next year",color="Orange")
plt.title("Metal price comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

## Pie Chart

```python
subjects = "Maths","Physcs","Chemistry","Englsih"
weightage= [15,15,15,10]

plt.pie(weightage,labels=subjects)
plt.show()
plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0))
plt.show()
plt.pie(weightage,labels=subjects,autopct='%1.1f%%')
plt.show()
plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),shadow=True)
plt.show()
plt.title("Subjects")
plt.pie(weightage,labels=subjects)
plt.show()
```

## Normal Distribution and Histogram

```python
xsn = np.random.randn(100)
sigma=5
u=60
x=np.round(xsn*sigma+u)
x2=np.round(xsn*5+40)
print(x)
```

```python
## Histogra,

plt.hist(x)
plt.hist(x2,alpha=0.8)
plt.xlabel("Marks")
plt.ylabel("Students")
plt.show()
```
