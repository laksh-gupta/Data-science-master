method to find local minima in any given function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

```python
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('F(x)')
plt.show()
```

```python
x = 0
lr =0.1
error = []
plt.plot(X,Y)
for i in range(50):
    grad=2*x-10 ##diff of fx wrt x
    x=x-lr*grad
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
    
```

```python
plt.plot(error)
```
