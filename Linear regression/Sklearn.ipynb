## Linear Regression using scikit-learn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=1000,n_features=2,n_informative=2,noise=10,random_state=1)
print(X.shape,y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(10,7))
ax=plt.axes(projection='3d')

ax.scatter3D(X[:,0],X[:,1],y,color='green');
plt.title('3D Scatter Plot')
plt.show()
```

```python
%matplotlib notebook
## to make the plot interactive
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
## Valuue of theta1 and theta 2 respectively 
```

```python
model.intercept_
## Value of theta0
```

```python
model.predict([X[0],X[1]])
```

```python
print(y[0],y[1])
```

```python
model.score(X,y)
```
